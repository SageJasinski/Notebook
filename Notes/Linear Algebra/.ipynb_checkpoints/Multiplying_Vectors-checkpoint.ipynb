# Multiplying Vectors


## Scaler multiplyer

When multiplying a *scaler* to a vector you simply multiply each entry of the vector to the scale value. This will give you a new *scaled* vector that will maintain its same direction but be scalled up or down in some way depending on the scale value.

$$
3 \begin{bmatrix}5\\8\end{bmatrix} = \begin{bmatrix}15\\24\end{bmatrix}
$$

As you can see when we take the scaler 3 and apply it to the vector $\begin{bmatrix}5\\8\end{bmatrix}$ we get $3*5 and 3*8$ giving us the new *scaled* vector of $\begin{bmatrix}15\\24\end{bmatrix}$.

Now if we plot the first vector on a graph with it's origin anywhere on the graph, for simplicity we will place the origin of the vector on the origin of the grid at (0,0), and over lay the new scaled vector (in blue) on the same graph we can see that the slope of the vector has staid the same but the length has scaled.

```python
import matplotlib.pyplot as plt

vectorA = [5,0]
originA = [8,0]

vectorB = [15,0]
originB = [24,0]


plt.plot(originB[0], vectorB[0], '>b')
plt.plot(originB, vectorB, '--b')
plt.plot(originA[0],vectorA[0], '>k')
plt.plot(originA,vectorA,'k')
plt.show
```
